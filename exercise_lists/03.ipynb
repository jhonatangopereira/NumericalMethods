```python
import numpy as np
import matplotlib.pyplot as plt
```

### Questão 1

```python
def bisection_method(a:float, b:float, func, iter:int=200, tol:float=1e-10):
    x = erro = counter = 1
    y_a = np.sign(func(a))
    y_b = np.sign(func(b))

    while erro > tol:
        if y_a * y_b > 0:
            print("Não existe raiz no intervalo.")
            return None
        elif y_a * y_b == 0:
            print(f"Raiz encontrada: {a, b}")
        
        previous_x = x
        x = (a + b) / 2
        y_x = func(x)

        if y_x == 0:
            print(f"Raiz encontrada: {y_x}")
            return None
        
        counter += 1
        if counter >= iter: break
        
        if y_a * y_x < 0:
            b = x
            y_b = y_x
        elif y_x * y_b < 0:
            a = x
            y_a = y_x
        erro = abs((x - previous_x) / x)
    return x
```

```python
func_q1 = lambda x: np.exp(-x) - 2
```

```python
bisection_method(-51, 50, func_q1)
```

```python
plt.plot(np.arange(-51, 50), [func_q1(x) for x in range(-51, 50)])
plt.grid(True)
plt.show()
```

```python
def bhaskara(a, b, c):
    delta = (b ** 2) - (4 * a * c)
    x1 = (-b + np.sqrt(delta)) / (2 * a)
    x2 = (-b - np.sqrt(delta)) / (2 * a)
    return x1, x2

print(bhaskara(-0.5, 2.5, 4.5))
```

```python
def false_position_method(a:float, b:float, func, iter:int=200, tol:float=1e-10):
    x = erro = counter = 1
    y_a = np.sign(func(a))
    y_b = np.sign(func(b))

    while erro > tol:
        if y_a * y_b > 0:
            print("Não existe raiz no intervalo.")
            return None
        elif y_a * y_b == 0:
            print(f"Raiz encontrada: {a, b}")
        
        previous_x = x
        x = (a + b) / 2
        x = b - ((y_b * (a - b)) / (y_a - y_b))
        y_x = func(x)

        if y_x == 0:
            print(f"Raiz encontrada: {y_x}")
            return None
        
        counter += 1
        if counter >= iter: break
        
        if y_a * y_x < 0:
            b = x
            y_b = y_x
        elif y_x * y_b < 0:
            a = x
            y_a = y_x
        erro = abs((x - previous_x) / x)
    return x
```
